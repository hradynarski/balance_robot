```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = np.loadtxt("./angle.log")
```

```python
target = data[:, 0]
sensed_angle = data[:, 1]
sensed_w =  data[:, 2]
dt =  data[:, 3]
```

```python
plt.plot(sensed_w, label='angular velocity')
plt.plot(sensed_angle, label='angle')
plt.legend(loc='best')
```

```python
def constrain(x, low, high):
    x = low if x<low else x
    x = high if x > high else x
    return x
    
```

```python
class PID():
    def __init__(self):
        self.dt = 0.01
        self.e_ = 0
        self.e__ = 0
        self.Kp_w = 6
        self.Kp = 0.4
        self.Ki = 20
        self.Kd = 0.
        self.MAX_CTRL_VAR = 210
        self.u = 0
        
    def cascade_control(self, target_angle, sensed_angle, sensed_w):

        target_w = (target_angle - sensed_angle)*self.Kp_w;

        e_w = target_w - sensed_w;
        self.u = self.u + (self.Kp + self.Ki*self.dt + self.Kd/self.dt)*e_w - \
                (self.Kp + 2*self.Kd/self.dt)*self.e_ + self.Kd/self.dt*self.e__;
	
        self.u = constrain(self.u, -self.MAX_CTRL_VAR, self.MAX_CTRL_VAR);

        self.e_ = e_w;
        self.e__ = e_w;
        return self.u;
```

```python
pid = PID()
control = []
for i  in xrange(len(data)):
    control.append(pid.cascade_control(target[i], sensed_angle[i], sensed_w[i]))
```

```python
plt.plot(control)
plt.plot(sensed_angle*10)
```
